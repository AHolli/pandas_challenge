```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
school_data_to_load = Path("C:/Users/mrshw/OneDrive/Desktop/Mod4_Starter_Code/PyCitySchools/Resources/schools_complete.csv")
student_data_to_load = Path("C:/Users/mrshw/OneDrive/Desktop/Mod4_Starter_Code/PyCitySchools/Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.
combined_df = pd.merge(school_data, student_data, how = "right", on="school_name") 
combined_df
```

## District Summary

```python
# Calculate the total number of unique schools
print(f'Number of Schools: {school_count}')
```

```python
# Calculate the total number of students
total_students = combined_df["Student ID"].unique().size
print(f'Total Students: {total_students}')
```

```python
# Calculate the total budget
total_budget = sum(combined_df["budget"].unique())
print(f'Total Budget: ${total_budget}')
```

```python
# Calculate the average (mean) math score
average_math_score = round(sum(combined_df["math_score"])/len(combined_df),2)
print(f'Average Math Score: {average_math_score}') 
```

```python
# Calculate the average (mean) reading score
average_reading_score = round(sum(combined_df["reading_score"])/len(combined_df),2)
print(f'Average Reading Score: {average_reading_score}')
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_grade = 70
number_passing = 0
math_scores = combined_df["math_score"]
for i in math_scores:
    if i >= passing_grade:
        number_passing = number_passing + 1
percent_passing_math = round(100*(number_passing/len(math_scores)),2)
print(f'% Passing Math: {percent_passing_math}')
number_passing
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_grade = 70
number_passing = 0
reading_scores = combined_df["reading_score"]
for i in reading_scores:
    if i >= passing_grade:
        number_passing = number_passing + 1
percent_passing_reading = round(100*(number_passing/len(reading_scores)),2) 
print(f'% Passing Reading: {percent_passing_reading}')
number_passing
```

```python
# Use the following to calculate the percentage of students that passed math and reading
total_passing = combined_df[(combined_df["math_score"] >= 70) & (combined_df["reading_score"] >= 70)].count()["student_name"]
total_percent_passing = round(100*(total_passing/(len(combined_df))),2)
print(f'% Passing both Math and Reading: {total_percent_passing}')
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Number of Unique Schools":[school_count],
    "Total Number of Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score":[average_math_score],
    "Average Reading Score":[average_reading_score],
    "% Passing Math":[percent_passing_math],
    "% Passing Reading":[percent_passing_reading],
    "% Overall Passing":[total_percent_passing]
})

district_summary["Total Number of Students"] = district_summary["Total Number of Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```
